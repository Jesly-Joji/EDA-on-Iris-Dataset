# **Visualization of count of Species**

```python
species_count=df["Species"].value_counts()
X=species_count.index
Y=species_count.values

sns.barplot(x=X,y=Y,palette="viridis")
plt.ylabel("Count")
```

# **Heatmap to visualize the correlation matrix of a dataset using Seaborn**

```python
c=df.corr(numeric_only=True)
sns.heatmap(c,annot=True,cmap="Blues")
```

# **Boxplot to compare the distributions of petal lengths across different species in the Iris dataset**

```python
#Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='PetalLengthCm', data=df)

plt.title("Comparison of Petal Lengths Across Iris Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
```

# **Scatter plot with a regression line to analyze the relationship between petal length and petal width in the Iris dataset**

```python
df1=df[df["Species"]=="Iris-setosa"]
df2=df[df["Species"]=="Iris-versicolor"]
df3=df[df["Species"]=="Iris-virginica"]
```

```python
df1.head(2)
```

```python
X=df["PetalLengthCm"]
Y=df["PetalWidthCm"]
```

```python
sns.scatterplot(x=X,y=Y,hue=df["Species"])
```

```python
from sklearn.linear_model import LinearRegression
sns.scatterplot(x=X,y=Y,hue=df["Species"])
model=LinearRegression()
X=np.array(X).reshape(-1,1)
model.fit(X,Y)
pred=model.predict(X)
plt.plot(X,pred)
```

# **Scatter plot with a regression line to analyze the relationship between sepal length and sepal width in the Iris dataset**

```python

model1=LinearRegression()
model2=LinearRegression()
model3=LinearRegression()
```

```python
# Data for Iris-setosa
X1 = df1["SepalLengthCm"]
Y1 = df1["SepalWidthCm"]

X1_array = np.array(X1).reshape(-1, 1)  # Reshape for sklearn
model1.fit(X1_array, Y1)
pred1 = model1.predict(X1_array)

# Data for Iris-versicolor
X2 = df2["SepalLengthCm"]
Y2 = df2["SepalWidthCm"]

X2_array = np.array(X2).reshape(-1, 1)  # Reshape for sklearn
model2.fit(X2_array, Y2)
pred2 = model2.predict(X2_array)

# Data for Iris-virginica
X3 = df3["SepalLengthCm"]
Y3 = df3["SepalWidthCm"]

X3_array = np.array(X3).reshape(-1, 1)  # Reshape for sklearn
model3.fit(X3_array, Y3)
pred3 = model3.predict(X3_array)

# PLOTS
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Iris-setosa")
sns.scatterplot(x=X1, y=Y1, color="red")
plt.plot(X1, pred1, color='blue')

plt.subplot(1, 3, 2)
plt.title("Iris-versicolor")
sns.scatterplot(x=X2, y=Y2, color="orange")
plt.plot(X2, pred2, color='blue')

plt.subplot(1, 3, 3)
plt.title("Iris-virginica")
sns.scatterplot(x=X3, y=Y3, color="green")
plt.plot(X3, pred3, color='blue')
```
